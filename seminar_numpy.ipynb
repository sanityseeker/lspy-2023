__Задача 2__: Даны два вектора x и y. Проверить, задают ли они одно и то же мультимножество (множество с повторяющимися элементами)
 
 
Например, для `x = np.array([3, 2, 2, 5])`, `y = np.array([5, 2, 3, 2])` ответ -- True

````python
x = np.array([3, 2, 2, 5, 7])
y = np.array([5, 2, 3, 2, 7])
````

````python
np.unique(x, return_counts=True)
````

````python
np.unique(y, return_counts=True)
````

````python
mask = np.sort(x) == np.sort(y)
````

````python
np.all(mask)
````

````python
np.all(np.sort(x) == np.sort(y))
````

Или

````python
np.array_equal(np.sort(x), np.sort(y))
````

Если у нас float

````python
a_flt = np.random.rand(10)
````

````python
a_flt
````

````python
b_flt = a_flt.copy()
````

````python
b_flt
````

````python
np.allclose(np.sort(a_flt), np.sort(b_flt))
````

Как можно ответить на этот вопрос без помощи numpy?

````python
diff = [round(i - j, 3) for i, j in zip(a_flt, b_flt)]
diff
````

````python
any(diff) is False
````

````python
from collections import Counter

a_cnts = Counter(round(i, 3) for i in a_flt)
b_cnts = Counter(round(i, 3) for i in b_flt)
````

````python
a_cnts, b_cnts
````

````python
a_cnts == b_cnts
````
